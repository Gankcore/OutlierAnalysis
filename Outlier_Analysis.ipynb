```python
# Import needed packages
import numpy as np
import pandas as pd
from openpyxl import load_workbook

# Create a DataFrame from the Excel file
df = pd.read_excel('raw_data.xlsx', skipinitialspace=True, inplace=True)

count = 'count'
dim = 'percent'
dimensionname = 'name'

# Outlier Calculation
if (df[dim].mean() + 3 * df[dim].std()) >1:
    A = 1
else:
    A = (df[dim].mean() + 3 * df[dim].std())
B = np.round(A,4)
BPercent = B * 100

# Create a DataFrame with only Outliers
df1 = df.loc[df[dim] >= B, ['code_to_identify',count,dim]]

# Create a DataFrame with Outliers and only 500 count or more
df2 = df1.loc[df1[count] >= 500, ['code_to_identify',count,dim]]
final_df = df2.sort_values(by=[dim], ascending=False)

BCount = final_df.dim.count()
print ("The ",dimensionname," Benchmark is ",BPercent,"% and there are ",BCount," ",dimensionname," Outliers.")

# Write DataFrame to a new Excel file
writer = pd.ExcelWriter('Outliers.xlsx', engine='xlsxwriter')
final_df.to_excel(writer, sheet_name=dimensionname)
writer.save()

count = 'count2'
dim = 'percent2'
dimensionname = 'name2'

# Outlier Calculation
if (df[dim].mean() + 3 * df[dim].std()) >1:
    A = 1
else:
    A = (df[dim].mean() + 3 * df[dim].std())
B = np.round(A,4)
BPercent = B * 100

# Create a DataFrame with only Outliers
df1 = df.loc[df[dim] >= B, ['code_to_identify',count,dim]]

# Create a DataFrame with Outliers and only 500 count or more
df2 = df1.loc[df1[count] >= 500, ['code_to_identify',count,dim]]
final_df = df2.sort_values(by=[dim], ascending=False)

BCount = final_df.dim.count()
print ("The ",dimensionname," Benchmark is ",BPercent,"% and there are ",BCount," ",dimensionname," Outliers.")

# Write DataFrame to an existing Excel file
book = load_workbook('Outliers.xlsx')
writer = pd.ExcelWriter('Outliers.xlsx', engine='openpyxl') 
writer.book = book
writer.sheets = dict((ws.title, ws) for ws in book.worksheets)
final_df.to_excel(writer, dimensionname)
writer.save()
```
